```python
#importing the libraries
import pandas as pd
import numpy as np
```

## Loading data into dataframe from url

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
df=pd.read_html(url, header=0)[0]

df.head(10)
```

## Dropping the rows with not assigned Borough

```python
df.drop(df[df.Borough == "Not assigned"].index, inplace=True)
```

## Final Dataset

```python
df.reset_index(drop=True,inplace=True)
df["Neighborhood"]=df["Neighborhood"].str.replace("/",",")
df.sort_values("Postal code", axis = 0, ascending = True, inplace = True,) 
df.head(20)
```

## Reading coordinates data into Dataframe and appending it to our df

```python
x=pd.read_csv("Geospatial_Coordinates.csv")
```
